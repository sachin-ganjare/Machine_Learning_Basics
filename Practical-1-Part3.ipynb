```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,5,7])
plt.title("Untitled Plot")
plt.ylabel("label-y")
plt.xlabel("label-x")
plt.show()
```

```python
x=[1,5,7]
y=[55,2,12]
x2=[85,12,42]
y2=[5,12,472]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
# using styles
from matplotlib import style
style.use('ggplot')
x=[1,5,7]
y=[55,2,12]
x2=[85,12,42]
y2=[5,12,472]
plt.plot(x,y,label='second')
plt.plot(x2,y2,label='first')
plt.legend()
plt.show()
```
